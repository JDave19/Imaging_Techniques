```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compute_enhancement_factor(air, m, z):
    ratio = z / (255.0 * m)
    return 1 + np.power(ratio * (255-air), 4)

# X-axis: air values (0 = bright, 255 = dark)
air_vals = np.linspace(0, 255, 256)

# Try different (z, m) settings
settings = [
    (0.9, 0.6), # high enhancement
    (0.6, 0.9), # low enhancement
    (1.1, 0.6), # Very high enhancement
]

plt.figure(figsize=(10, 6))

for z, m in settings:
    factor = compute_enhancement_factor(air_vals, m, z)
    print(f"Min: {factor.min()}, Max: {factor.max()}")
    plt.plot(air_vals, factor, label=f"z={z}, m={m}")

plt.title("Enhancement Factor vs Air Value")
plt.xlabel("air[idx] (255 - min(R,G,B))")
plt.ylabel("Enhancement Factor")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
